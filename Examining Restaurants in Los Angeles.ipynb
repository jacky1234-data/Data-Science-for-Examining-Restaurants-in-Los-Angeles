## Set the serach query as restaurant, as we want to know the distribution of restaurants

```python
search_query = 'Restaurant'
radius = 15000   ## Radius is 15km
print(search_query + ' .... OK!')
```

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&oauth_token={}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude,ACCESS_TOKEN, 20180604, search_query, radius, LIMIT)
url
```

```python
results = requests.get(url).json()
results
```

```python
response = results['response']['venues']
response
```

```python
response_df = pd.DataFrame(columns = ['Name','Address','Full Address','Latitude','Longitude','Categories','Distance From Origin'])
response_df
```

```python
for item in response:
    try:
        Name = item['name']
        Address = item['location']['address']
        Full_Address = item['location']['formattedAddress'][0] + " " + item['location']['formattedAddress'][1]
        Latitude = item['location']['lat']
        Longitude = item['location']['lng']
        Categories = item['categories'][0]['shortName']
        Distance_From_Origin = item['location']['distance']
    except KeyError:
        Name = item['name']
        Address = 'Not Found'
        Full_Address = 'Not Found'
        Latitude = item['location']['lat']
        Longitude = item['location']['lng']
        Categories = item['categories'][0]['shortName']
        Distance_From_Origin = item['location']['distance']
    except IndexError:
        Name = item['name']
        Address = item['location']['address']
        Latitude = item['location']['lat']
        Longitude = item['location']['lng']
        Categories = 'Others'
        Distance_From_Origin = item['location']['distance']
    response_df = response_df.append({'Name':Name,'Address':Address, 'Full Address':Full_Address,'Latitude':Latitude,'Longitude':Longitude,'Categories':Categories,'Distance From Origin':Distance_From_Origin},ignore_index=True)
response_df
    
```

```python
response_df.shape
```

```python
indexNames = response_df[response_df['Address'] == 'Not Found'].index
 
# Delete these row indexes from dataFrame
response_df.drop(indexNames , inplace=True)
response_df
```

```python
response_df = response_df.reset_index(drop = True)
response_df
```

```python
response_df['Categories'].value_counts()
```

```python
len(response_df['Categories'].unique())
```

```python
df_analysis = response_df[['Latitude','Longitude','Categories','Distance From Origin']]
df_analysis = pd.concat([df_analysis,pd.get_dummies(df_analysis['Categories'])], axis=1)
df_analysis
```

```python
df_analysis = df_analysis.drop(['Categories'], axis = 1)
df_analysis
```

```python
kclusters = 7
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_analysis)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
len(kmeans.labels_)
```

```python
response_df['Cluster Label'] = kmeans.labels_
response_df
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(response_df['Latitude'], response_df['Longitude'], response_df['Categories'], response_df['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
